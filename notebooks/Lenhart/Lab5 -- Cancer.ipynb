In this Lab, we modelize the growth of a tumor under chemotherapy (control) using a model that follows Skipper's log-kill hypothesis. That is:
$$
\min\int_0^T aN(t)² + u(t)² \\
\mathrm{subject\: to}\quad N'(t) = rN(t)\ln\Big( \frac{1}{N(t)} \Big) - u(t)\delta N(t)\\
N(0)=N_0\, ,\; u(t)\geq0
$$

##### Parameters
r : growth rate of the tumor

$\delta$ : magnitude of the dose

a : weight paramater; control the importance of minimizing $N(t)²$ vs $u(t)²$
